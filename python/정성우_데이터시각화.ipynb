```python
import pandas as pd
import numpy as np
import sys
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
total=pd.read_excel('C:/Users/정성우/Downloads/전체.xlsx')
```

```python
#Reporting Economy를 index로 
total.index = total['Reporting Economy']

total
```

```python
#첫 3년, 마지막 3년 데이터가 있는 국가만 골라서 테이블 만들기
total_invalid = total[(np.isnan(total['1990'])&np.isnan(total['1991'])&np.isnan(total['1992']))
                   |(np.isnan(total['2017'])&np.isnan(total['2018'])&np.isnan(total['2019']))]
total_valid= total.drop(total_invalid.index)
total_valid
```

```python
N_resource = pd.read_excel('C:/Users/정성우/Downloads/원자재.xlsx')
```

```python
#Reporting Economy를 index로 
N_resource.index = N_resource['Reporting Economy']

N_resource
```

```python
#첫 3년, 마지막 3년 데이터가 있는 국가만 골라서 테이블 만들기
invalid = N_resource[(np.isnan(N_resource['1990'])&np.isnan(N_resource['1991'])&np.isnan(N_resource['1992']))
                   |(np.isnan(N_resource['2017'])&np.isnan(N_resource['2018'])&np.isnan(N_resource['2019']))]
N_valid= N_resource.drop(invalid.index)
N_valid
```

```python
#필요한 값이 있는 열이 있는 곳만 선택
#N_valid와 total_valid에 모두 포함되는 국가만을 뽑기 위해 교집합 연산 np.intersect1d 사용
N_valid = N_valid.loc[np.intersect1d(N_valid.index, total_valid.index), '1990':'2019']

N_valid
```

```python
#전체 교역에서 valid 한 국가의 값만 추출
#N_valid와 total_valid에 모두 포함되는 국가만을 뽑기 위해 교집합 연산 np.intersect1d 사용
total_valid = total_valid.loc[np.intersect1d(N_valid.index, total_valid.index), '1990':'2019']
total_valid
```

```python
#비율 구하기
NR_ratio = N_valid/total_valid
NR_ratio
```

```python
#NaN을 무시하고 첫 3년, 마지막 3년의 평균 구하기

#1990~1992년 평균 구하기
masked_data_909192 = np.ma.masked_array(NR_ratio.loc[:,'1990':'1992'], np.isnan(NR_ratio.loc[:,'1990':'1992']))
weights = [1, 1, 1]
average = np.ma.average(masked_data_909192, axis=1, weights=weights)
result1 = average.filled(np.nan)
result1

#1990~1992년 평균을 열로 삽입
NR_ratio['1990~1992 average']=result1


#2017~2019년 평균 구하기
masked_data_171819 = np.ma.masked_array(NR_ratio.loc[:,'2017':'2019'], np.isnan(NR_ratio.loc[:,'2017':'2019']))
weights = [1, 1, 1]
average = np.ma.average(masked_data_171819, axis=1, weights=weights)
result2 = average.filled(np.nan)
result2

#2017~2019년 평균을 열로 삽입
NR_ratio['2017~2019 average']=result2

NR_ratio
```

```python
#2017~2019년의 평균에서 1990~1992 평균을 뺀 값 구하기

NR_ratio['ratio gap'] = NR_ratio['2017~2019 average']-NR_ratio['1990~1992 average']

#1990~1992 평균 대비 2017~2019년 평균의 성장률
NR_ratio['ratio growth rate(%)'] = (NR_ratio['2017~2019 average']-NR_ratio['1990~1992 average'])/NR_ratio['1990~1992 average']*100

NR_ratio
```

## HDI 처리하기

```python
#HDI 읽어오기

HDI=pd.read_excel('C:/Users/정성우/Downloads/Human Development Index (HDI).xlsx')
```

```python
HDI
```

```python
# ...으로 표기된 값을 NaN으로 치환
HDI = HDI.replace("..", np.NaN)
HDI
```

```python
#index를 country로 바꾸고 country 열은 drop
HDI.index = HDI['Country']
HDI=HDI.drop('Country', axis=1)
```

```python
#첫 3년, 마지막 3년 데이터가 있는 국가만 골라서 테이블 만들기
HDI_invalid = HDI[(np.isnan(HDI[1990])&np.isnan(HDI[1991])&np.isnan(HDI[1992]))
                   |(np.isnan(HDI[2017])&np.isnan(HDI[2018])&np.isnan(HDI[2019]))]
HDI_valid= HDI.drop(HDI_invalid.index)
HDI_valid
```

```python
#NaN을 무시하고 첫 3년, 마지막 3년의 평균 구하기

#1990~1992년 평균 구하기
masked_data_909192 = np.ma.masked_array(HDI_valid.loc[:,[1990,1991,1992]], np.isnan(HDI_valid.loc[:,[1990,1991,1992]]))
weights = [1, 1, 1]
average = np.ma.average(masked_data_909192, axis=1, weights=weights)
result1 = average.filled(np.nan)
result1

#1990~1992년 평균을 열로 삽입
HDI_valid['1990~1992 average']=result1


#2017~2019년 평균 구하기
masked_data_171819 = np.ma.masked_array(HDI_valid.loc[:,[2017, 2018, 2019]], np.isnan(HDI_valid.loc[:,[2017,2018,2019]]))
weights = [1, 1, 1]
average = np.ma.average(masked_data_171819, axis=1, weights=weights)
result2 = average.filled(np.nan)
result2

#2017~2019년 평균을 열로 삽입
HDI_valid['2017~2019 average']=result2

HDI_valid
```

```python
#2017~2019년의 평균에서 1990~1992 평균을 뺀 값 구하기

HDI_valid['ratio gap'] = HDI_valid['2017~2019 average']-HDI_valid['1990~1992 average']

#1990~1992 평균 대비 2017~2019년 평균의 성장률
HDI_valid['ratio growth rate(%)'] = (HDI_valid['2017~2019 average']-HDI_valid['1990~1992 average'])/HDI_valid['1990~1992 average']*100

#처리된 HDI_valid 이름을 HDI로 변경
HDI=HDI_valid

#HDI의 index에 발생한 띄어쓰기 제거
HDI.index = HDI.index.str.strip()
```

```python
HDI
```

```python
#Sub-Saharan 국가들만 모아보기
SubSahara = ["Angola", "Benin", "Botswana", "Burkina Faso", "Burundi", "Cabo Verde", "Cameroon", 
             "Central African Republic", "Chad", "Comoros", "Congo", "Côte d'Ivoire", "Djibouti", 
             "Equatorial Guinea", "Eritrea", "Eswatini", "Ethiopia","Ethiopia (+ Eritrea)", "Gabon", 
             "The Gambia", "Ghana", "Guinea", "Guinea-Bissau", "Kenya", "Lesotho", "Liberia", "Madagascar", 
             "Malawi", "Mali", "Mauritania", "Mauritius", "Mozambique", "Namibia", "Niger", "Nigeria", "Rwanda", 
             "Sao Tomé and Principe", "Senegal", "Seychelles", "Sierra Leone", "Somalia", "South Africa", "Sudan", 
             "Tanzania", "Togo", "Uganda", "Zambia", "Zimbabwe"]

#Sub-Saharan 국가 중 자원의존률을 구한 국가
NR_SS_list = []

for country in NR_ratio.index:
    if country in SubSahara:
        NR_SS_list.append(country)

NR_SS_list
```

```python
HDI.index
```

```python
#Sub-Saharan 국가 중 유효한 HDI가 있는 국가
HDI_SS_list = []

for country in HDI.index:
    if country in SubSahara:
        HDI_SS_list.append(country)
        
HDI_SS_list
```

```python
#NR_SS_list와 HDI_SS_list 모두에 속한 국가 뽑기

NH_SS_list = []

for country in NR_SS_list:
    if country in HDI_SS_list:
        NH_SS_list.append(country)

NH_SS_list
```

유효한 원자재 의존도와 HDI가 있는 NH_SS_list 국가들을 중심으로 가설 검정

```python
#NH_SS_list 국가들의 원자재 의존도확인

SS_NR = NR_ratio.loc[NH_SS_list].iloc[:,-4:]
SS_NR
```

```python
SS_HDI = HDI.loc[NH_SS_list].iloc[:,-4:]
SS_HDI
```

```python
#통합된 df 만들기

#통합전 동일한 column 이름 구분 
SS_NR.columns = ['NR_1990~1992 average', 'NR_2017~2019 average', 'NR_ratio gap', 'NR_ratio growth rate(%)']
SS_HDI.columns = ['HDI_1990~1992 average', 'HDI_2017~2019 average', 'HDI_ratio gap', 'HDI_ratio growth rate(%)']

SS_join = pd.merge(SS_NR, SS_HDI, left_on=SS_NR.index,right_on=SS_HDI.index, how = 'inner')

SS_join
```

```python
#산포도 시각화 
SS_join.plot.scatter('NR_1990~1992 average', 'HDI_1990~1992 average', title = '1990~1992 Sub-Saharan Natural resource ratio X HDI')
SS_join.plot.scatter('NR_2017~2019 average', 'HDI_2017~2019 average', title = '2017~2019 Sub-Saharan Natural resource ratio X HDI')
SS_join.plot.scatter('NR_ratio gap', 'HDI_ratio gap', title = '1990~2019 Sub-Saharan Natural resource ratio gap X HDI gap')
```

## 사하라 이남 국가와 비교할 벤치마크 국가 선정

벤치마크 선정 기준
- 1990-1992년 자원의존도보다 2017-2019 자원의존도가 0.03 이상 하락한 국가(ratio gap이 0.03이하인 것이 기존 경제에서 자원의존도가 있는 국가가 유의미하게 비중이 감소한 것으로 판단함)
- 1990~1992년 HDI가 0.7 이하인 국가(0.63이 사하라 이남 국가 중 최대값)

```python
#1990-1992년 자원의존도보다 2017-2019 자원의존도가 0.05 이상 하락한 국가 찾기
BM_NR = NR_ratio[NR_ratio['ratio gap'] <= -0.03]

#필요한 열만 선택
BM_NR = BM_NR.iloc[:,-4:]
BM_NR
```

```python
#1990~1992년 HDI가 0.63(사하라 이남 국가 중 최대값) 이하인 국가

BM_HDI = HDI[HDI['1990~1992 average'] < 0.7]

#필요한 열만 선택
BM_HDI = BM_HDI.iloc[:,-4:]
BM_HDI
```

```python
#두 테이블의 겹치는 열 이름 수정
BM_NR.columns = ['NR_1990~1992 average', 'NR_2017~2019 average', 'NR_ratio gap', 'NR_ratio growth rate(%)']
BM_HDI.columns = ['HDI_1990~1992 average', 'HDI_2017~2019 average', 'HDI_ratio gap', 'HDI_ratio growth rate(%)']

#두 국가의 교집합 구하기 
BM_join = pd.merge(BM_NR, BM_HDI, left_on=BM_NR.index,right_on=BM_HDI.index, how = 'inner')

BM_join
```

```python
#사하라 이남 국가에 포함되는 국가 제외하기
#BM_join = BM_join.drop(
BM_join = BM_join[BM_join['key_0'] != 'Kenya']
BM_join = BM_join[BM_join['key_0'] != 'Togo']
BM_join
```

```python
BM_join['group']='Benchmark'
BM_join
```

```python
SS_join['group']='SubSaharan'
```

```python
#산포도를 그리기 위한 테이블 병합
result = pd.concat([SS_join, BM_join])
```

```python
result.index = range(0,21)
result
```

```python
sns.scatterplot( x = 'NR_1990~1992 average',
               y = 'HDI_1990~1992 average',
               hue = 'group',
               style = 'group',
               s=100, 
               data = result)

plt.title('1990~1992 Natural resource ratio X HDI')
plt.show
```

```python
sns.scatterplot( x = 'NR_2017~2019 average',
               y = 'HDI_2017~2019 average',
               hue = 'group',
               style = 'group',
               s=100, 
               data = result)

plt.title('2017~2019 Natural resource ratio X HDI')
plt.show
```

벤치마크 국가들은 자원의존도 하락폭이 클 수록 HDI가 적게 상승한 것으로 보이나, 사하라 이남 국가들은 자원의존도 하락폭이 커지는 것과 HDI가 상관이 없거나 미세한 음의 상관관계를 갖는 것으로 보임

## 위치와 상관없이 원자재에 대한 무역의존도와 HDI 추이 비교

사하라 이남으로 국가를 한정하고 비교한 경우 국가의 모수가 적어 유의미한 결과를 얻기 어려움
따라서 아래 기준에 맞는 국가들을 골라 원자재 의존도와 HDI의 상관관계 분석

기준
- 1990~1992 HDI 평균 0.7 이하
- 1990~1992 자원의존도 평균 0.1 이상


```python
#NR_ratio에서 필요한 열만 선택하기

#1990-1992년 자원의존도 평균이 0.1 이상인 국가 찾기
NR_ratio_1 = NR_ratio[NR_ratio['1990~1992 average'] >= 0.1]

#필요한 열만 선택
NR_ratio_1 = NR_ratio_1.iloc[:,-4:]
NR_ratio_1
```

```python
#HDI에서 필요한 열만 선택하기

#1990-1992년 HDI 평균이 0.7 이하인 국가 찾기
HDI_1 = HDI[HDI['1990~1992 average'] <= 0.7]

#필요한 열만 선택
HDI_1 = HDI_1.iloc[:,-4:]
HDI_1
```

```python
#두 테이블의 겹치는 열 이름 수정
NR_ratio_1.columns = ['NR_1990~1992 average', 'NR_2017~2019 average', 'NR_ratio gap', 'NR_ratio growth rate(%)']
HDI_1.columns = ['HDI_1990~1992 average', 'HDI_2017~2019 average', 'HDI_ratio gap', 'HDI_ratio growth rate(%)']

#테이블의 교집합 국가 구하기 
country_merged = pd.merge(NR_ratio_1, HDI_1, left_on=NR_ratio_1.index,right_on=HDI_1.index, how = 'inner')

country_merged
```

```python
country_merged.plot.scatter('NR_1990~1992 average', 'HDI_1990~1992 average', title = '1990~1992 Natural resource ratio X HDI')
country_merged.plot.scatter('NR_2017~2019 average', 'HDI_2017~2019 average', title = '2017~2019  Natural resource ratio X HDI')
```

```python
sns.scatterplot( x = 'NR_ratio gap',
               y = 'HDI_ratio gap',
               s=100, 
               data = country_merged)
sns.regplot(x='NR_ratio gap', y='HDI_ratio gap', lowess=True, data=country_merged)

plt.title('1990~2019 Natural resource ratio gap X HDI gap')
plt.show
```

```python
sns.scatterplot( x = 'NR_ratio growth rate(%)',
               y = 'HDI_ratio growth rate(%)',
               s=100, 
               data = country_merged)

sns.regplot(x='NR_ratio growth rate(%)', y='HDI_ratio growth rate(%)', lowess=True, data=country_merged)


plt.title('1990~2019 Natural resource ratio growth rate(%) X HDI growth rate(%)')
plt.show
```
