```python
import numpy as np
import pandas as pd

#DB가져오기
boonja = pd.read_excel("원자재.xlsx", engine="openpyxl")
boonmo = pd.read_excel("전체.xlsx", engine="openpyxl")
HDI = pd.read_excel("Human Development Index (HDI).xlsx", engine="openpyxl")
```

```python
#사하라 이남 아프리카 국가 목록
SubSahara = ["Angola", "Benin", "Botswana", "Burkina Faso", "Burundi", "Cabo Verde", "Cameroon", 
             "Central African Republic", "Chad", "Comoros", "Congo", "Côte d'Ivoire", "Djibouti", 
             "Equatorial Guinea", "Eritrea", "Eswatini", "Ethiopia","Ethiopia (+ Eritrea)", "Gabon", 
             "The Gambia", "Ghana", "Guinea", "Guinea-Bissau", "Kenya", "Lesotho", "Liberia", "Madagascar", 
             "Malawi", "Mali", "Mauritania", "Mauritius", "Mozambique", "Namibia", "Niger", "Nigeria", "Rwanda", 
             "Sao Tomé and Principe", "Senegal", "Seychelles", "Sierra Leone", "Somalia", "South Africa", "Sudan", 
             "Tanzania", "Togo", "Uganda", "Zambia", "Zimbabwe"]

#사하라 이남 국가들만 모아두기
SubJa = pd.DataFrame()
#사하라 이남 국가들만 모아두기
noSubJa = pd.DataFrame()
#사하라 이남 국가들만 모아두기
SubMo = pd.DataFrame()
#사하라 이남 국가들만 모아두기
noSubMo = pd.DataFrame()
```

```python
#자료 정리 - boonja1으로 사용
boonja1 = boonja.drop([0], axis=0)
rename_col = np.array(boonja.iloc[1])
rename_col[3:] = np.array(rename_col[3:], dtype=np.int16)
boonja1.columns = rename_col
```

```python
#자료 정리 - boonmo1으로 사용
boonmo1 = boonmo.drop([0], axis=0)
rename_col = np.array(boonmo.iloc[1])
rename_col[3:] = np.array(rename_col[3:], dtype=np.int16)
boonmo1.columns = rename_col
```

```python
#전체 국가 숫자
country_num = len(boonja1)
country_num1 = len(boonmo1)
# print(country_num==country_num)
```

```python
boonja1 = boonja1.drop(['Product/Sector', 'Partner Economy'], axis=1)
boonmo1 = boonmo1.drop(['Product/Sector', 'Partner Economy'], axis=1)
```

```python
#사하라이남국가와 아닌 국가로 df 나누기(분자)
for i in range(1, country_num):
    if boonja1.iloc[i]["Reporting Economy"] in SubSahara:
        SubJa = SubJa.append(boonja1.iloc[i], ignore_index=True)
    else:
        noSubJa = noSubJa.append(boonja1.iloc[i], ignore_index=True)
```

```python
#사하라이남국가와 아닌 국가로 df 나누기(분모)
for i in range(1, country_num):
    if boonmo1.iloc[i]["Reporting Economy"] in SubSahara:
        SubMo = SubMo.append(boonmo1.iloc[i], ignore_index=True)
    else:
        noSubMo = noSubMo.append(boonmo1.iloc[i], ignore_index=True)
```

```python
a = np.array(SubJa["Reporting Economy"])
b = np.array(SubMo["Reporting Economy"])
print(a)
print(b)
```

```python
k = np.intersect1d(a,b)
print(k)
```

```python
temp_df1 = pd.DataFrame()
for i in range(len(a)):
    if SubJa.iloc[i]["Reporting Economy"] in k:
        temp_df1 = temp_df1.append(SubJa.iloc[i], ignore_index=True)
temp_df2 = pd.DataFrame()
for i in range(len(b)):
    if SubMo.iloc[i]["Reporting Economy"] in k:
        temp_df2 = temp_df2.append(SubMo.iloc[i], ignore_index=True)
```
