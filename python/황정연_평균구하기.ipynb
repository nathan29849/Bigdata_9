```python
import numpy as np
import pandas as pd

#DB가져오기
subsa = pd.read_excel("의존율_사하라이남.xlsx", engine="openpyxl")
not_subsa = pd.read_excel("의존율_사하라이남_제외.xlsx", engine="openpyxl")
```

```python
new_col = np.array(["Country"])
new_col = np.append(new_col, np.arange(1990,2020))

subsa.columns = new_col
not_subsa.columns = new_col
```

```python
country_groups = [subsa, not_subsa]
```

```python
ladder_exists1 = pd.DataFrame()
ladder_exists2 = pd.DataFrame()
```

```python
for i in range(len(subsa)):
    if (subsa.iloc[i]["1990"]>=40) or (subsa.iloc[i]["1991"]>=40) or (subsa.iloc[i]["1992"]>=40) or (subsa.iloc[i]["1993"]>=40 or (subsa.iloc[i]["1994"]>=40) or (subsa.iloc[i]["1995"]>=40) or (subsa.iloc[i]["1996"]>=40) or (subsa.iloc[i]["1997"]>=40) or (subsa.iloc[i]["1998"]>=40) or (subsa.iloc[i]["1999"]>=40)):
        ladder_exists1 = ladder_exists1.append(subsa.iloc[i], ignore_index=True)
```

```python
for i in range(len(not_subsa)):
    if (not_subsa.iloc[i]["1990"]>=40) or (not_subsa.iloc[i]["1991"]>=40) or (not_subsa.iloc[i]["1992"]>=40) or (not_subsa.iloc[i]["1993"]>=40 or (not_subsa.iloc[i]["1994"]>=40) or (not_subsa.iloc[i]["1995"]>=40) or (not_subsa.iloc[i]["1996"]>=40) or (not_subsa.iloc[i]["1997"]>=40) or (not_subsa.iloc[i]["1998"]>=40) or (not_subsa.iloc[i]["1999"]>=40)):
        ladder_exists2 = ladder_exists2.append(not_subsa.iloc[i], ignore_index=True)
```

```python
ladder_exists1.to_excel("의존율 높은 사하라 이남.xlsx")
ladder_exists2.to_excel("의존율 높은 사하라 이남 제외.xlsx")
```

```python
ladder_exists1_means = pd.DataFrame()
ladder_exists2_means = pd.DataFrame()
```

```python
for i in range(len(ladder_exists1)):
    temp_sum_front = 0
    temp_boonmo_front = 0
    temp_sum_rear = 0
    temp_boonmo_rear = 0
    for j in range(10):
        front = str(1990+j)
        curr_val_front = ladder_exists1.iloc[i][front]
        rear = str(2000+j)
        curr_val_rear = ladder_exists1.iloc[i][rear]
        if (curr_val_front > 0) or (curr_val_front < 0):
            temp_sum_front += curr_val_front
            temp_boonmo_front += 1
        if (curr_val_rear > 0) or (curr_val_rear < 0):
            temp_sum_rear += curr_val_rear
            temp_boonmo_rear += 1
    print(temp_sum_front, temp_boonmo_rear)
    try:
        temp_front_mean = round(temp_sum_front/temp_boonmo_front, 2)
    except:
        temp_front_mean = "NA"
    try:
        temp_rear_mean = round(temp_sum_rear/temp_boonmo_rear, 2)
    except:
        temp_rear_mean = "NA"
    temp_series = pd.Series([ladder_exists1.iloc[i]["Country"], temp_front_mean, temp_rear_mean])
    ladder_exists1_means = ladder_exists1_means.append(temp_series, ignore_index = True)

for i in range(len(ladder_exists2)):
    temp_sum_front = 0
    temp_boonmo_front = 0
    temp_sum_rear = 0
    temp_boonmo_rear = 0
    for j in range(10):
        front = str(1990+j)
        curr_val_front = ladder_exists2.iloc[i][front]
        rear = str(2000+j)
        curr_val_rear = ladder_exists2.iloc[i][rear]
        if (curr_val_front > 0) or (curr_val_front < 0):
            temp_sum_front += curr_val_front
            temp_boonmo_front += 1
        if (curr_val_rear > 0) or (curr_val_rear < 0):
            temp_sum_rear += curr_val_rear
            temp_boonmo_rear += 1
    print(temp_sum_front, temp_boonmo_rear)
    try:
        temp_front_mean = round(temp_sum_front/temp_boonmo_front, 2)
    except:
        temp_front_mean = "NA"
    try:
        temp_rear_mean = round(temp_sum_rear/temp_boonmo_rear, 2)
    except:
        temp_rear_mean = "NA"
    temp_series = pd.Series([ladder_exists2.iloc[i]["Country"], temp_front_mean, temp_rear_mean])
    ladder_exists2_means = ladder_exists2_means.append(temp_series, ignore_index = True)
```
